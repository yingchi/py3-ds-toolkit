```python
from yellowbrick.datasets import load_credit
from yellowbrick.features import PCA 
```

```python
X, y = load_credit()
classes = ['account in defaut', 'current with bills']
```

```python
visualizer = PCA(scale=True, classes=classes)
visualizer.fit_transform(X, y)
visualizer.show()
```
